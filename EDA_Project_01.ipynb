 **`Contains 488 missing values`**
- **agent**:

 **`Contains 16,340 missing values`**

 **`Contains 112,593 missing values, which is approximately 94% of the data in this column`**


📍 Based on the nature of each column and the number of missing values, we have decided on the following methods:


-  **"children" and "country" Columns** :


    Missing values will be filled using the 'Mode' (most frequent value)
    
    which is suitable for categorical data and provides a stable representation of the data.

- **"agent" Column** :


    We will fill missing values with the 'Mean' (average value) as it represents the overall distribution
    
    of this numerical column without biasing toward any particular category
    
    This approach minimizes the impact on analysis accuracy.

- **"company" Column** :


    This column will be dropped entirely from the analysis due to the high proportion of missing values (about 94%)
    
    Filling in these missing values could negatively impact the final analysis results.



````python
data["children"] = data["children"].fillna(data["children"].mode()[0])
data["country"] = data["country"].fillna(data["country"].mode()[0])
data["agent"]=data["agent"].fillna(data["agent"].mean())
````

## **Step 2: Identifying Outliers Using IQR**

**Overview**  

    In this step, we will display the outliers that we wish to remove from the dataset.

**Methodology**  

    We will use the **Interquartile Range (IQR)** method,
    which is an effective way to identify outliers based on the distribution of the data.

**Purpose**

    This approach will help us recognize values that deviate from the overall pattern of the data,
    allowing us to improve the quality of the data we will analyze.

---



````python
def Handling_Outlier(ColName):
  Q1 = np.quantile(ColName, 0.25)
  Q3 = np.quantile(ColName, 0.75)
  IQR = Q3-Q1
  Lower = Q1 - 1.5*IQR
  Upper = Q3 + 1.5*IQR
  outlier = ColName [ (ColName < Lower) | (ColName > Upper) ]
  return outlier
````

````python
adr_outliers = Handling_Outlier(num['lead_time'])
print(adr_outliers)
````

### Documentation for Converting the Date Column to Numerical Features

#### 1. **Step: Converting the Date Column**
**Reason**: The column `reservation_status_date` contains dates, and since machine learning models require numerical inputs, we will convert this column into numerical features such as year, month, and day. This will help improve the performance of the analysis model.

#### 2. **Features to be Extracted**
- **Year**: Extract the year from the date.
- **Month**: Extract the month from the date.
- **Day**: Extract the day from the date.
- **Week Number**: Optionally, extract the week number from the date if it is useful for your model.



````python
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'])

data['reservation_year'] = data['reservation_status_date'].dt.year
data['reservation_month'] = data['reservation_status_date'].dt.month
data['reservation_day'] = data['reservation_status_date'].dt.day

data.drop(columns=['reservation_status_date'], inplace=True)
````
